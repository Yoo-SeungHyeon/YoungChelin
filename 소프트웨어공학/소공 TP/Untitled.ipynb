```python
def cache_simul_repeat_gcc(size, assoc, repl):
    size = size
    for j in range(assoc): 
        as_num = 2**j
        for i in range(4):
            bsize = 2**(i+3)
            nsets = int(size/(bsize*as_num))
            inst = "./sim-cache -redir:sim cache{}_{}_{}_{}.txt -max:inst 1000000 -cache:il1 il1:{}:{}:{}:{} ../benchmark/gcc/gcc00.peak.ev6 ../benchmark/gcc/cp-decl.i".format(size,nsets,bsize,as_num,nsets,bsize,as_num,repl)
            system(inst)


cache_simul_repeat_gcc(8192,3,'l')
cache_simul_repeat_gcc(8192,3,'f')
cache_simul_repeat_gcc(8192,3,'r')
```
